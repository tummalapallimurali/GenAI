    * Average waiting times for jobs
    * Average Queue length.
    * Server Utilziation

  * system behaviour and its performance with reference to the jobs that are manufactured by the system,
  * considering their routings,

  * i.e. a power-queueing model.

  * ***average service times***, i.e. production times at the **different departments**.
  * are considered in the power-queueing model, too,
  * and the **average energy requirement** of the system can be computed

# According to the Kendall-Lee notation,

* the queuing model proposed for the power customers is a GI/G//GD// (an infinite-server system,
* in which **inter-arrival** and service times follow general probability distributions,
* and the **maximum allowed number of customers** in the system is infinite).
* The Queuing Theory approach allows the ***calculation of the probability*** that a given number of customers will be in the system,

* thus offering the possibility to calculate the service level obtained when setting a value to the contractual power level (i.e., the number of customers).



## Job Queueing Model:

* $_{i}(Job/h)$  - Arrival rate of Jobs, ith dept.
* $\mu_{i}(h)$ :Processing time of job at ith dept
* $_{i}(%)$ : Utilization coefficient = $_{i}/\mu_{i}$
* $L_{i}(Job) = _{i}/(\mu_{i}-_{i})$: Average number of jobs in ith dept.
* $W_{i}(h)= L_{i} / _{i} $ : Avg time spent by each job in each manufacturing dept. (Little law)

## Power Queueing Model:

From Energy use point of view:

* $_{p,i}(KW/h)$ - When a physical job is going to be processed at i dept , it can start production. must wait in queue

* $_{p,i}= P_{nom,i} * f_{l,i} * _{i} $

## Arrival rate, service rate and utilisation rate of the departments.


```python
'''
# Calcualte
'''

D1 = (0.25 + 0.25) * 15
D2 = (0.25 + 0.25+ 0.25) * 15
D3 = (0.25+ 0.25+ 0.25) * 15
D4 = (0.25+ 0.25+ 0.25+ 0.25) * 15

print("Lamda1(Job/h)", D1)
print("Lamda2(Job/h)", D2)
print("Lamda3(Job/h)", D3)
print("Lamda4(Job/h)", D4)

mu_hr = 0.05

utilization1 = D1 * mu_hr
utilization2 = D2 * mu_hr
utilization3 = D3 * mu_hr
utilization4 = D4 * mu_hr

print("Utilization1", utilization1)
print("Utilization2", utilization2)
print("Utilization3", utilization3)
print("Utilization4", utilization4)
```

## Arrival rate and service rate of M/M/ model.

```python
lamda_j = [7.5, 11.25, 11.25, 15.0]
P_nominal = 10
f_i = 0.8

# calulcate weighted sum of _nominal and lambda_j

lamda_i = []
for i in range(len(lamda_j)):
    lamda_i.append(lamda_j[i] * P_nominal* f_i)
    #sum of labda_i


sum_labda_i = sum(lamda_i)
print("Lamda_h (KWh/h)", sum_labda_i)
```
