# According to the Kendall-Lee notation,

* the queuing model proposed for the power customers is a GI/G//GD// (an infinite-server system,
* in which **inter-arrival** and service times follow general probability distributions,
* and the **maximum allowed number of customers** in the system is infinite).
* The Queuing Theory approach allows the ***calculation of the probability*** that a given number of customers will be in the system,

* thus offering the possibility to calculate the service level obtained when setting a value to the contractual power level (i.e., the number of customers).



## Job Queueing Model:

* $_{i}(Job/h)$  - Arrival rate of Jobs, ith dept.
* $\mu_{i}(h)$ :Processing time of job at ith dept
* $_{i}(%)$ : Utilization coefficient = $_{i}/\mu_{i}$
* $L_{i}(Job) = _{i}/(\mu_{i}-_{i})$: Average number of jobs in ith dept.
* $W_{i}(h)= L_{i} / _{i} $ : Avg time spent by each job in each manufacturing dept. (Little law)

## Power Queueing Model:

From Energy use point of view:

* $_{p,i}(KW/h)$ - When a physical job is going to be processed at i dept , it can start production. must wait in queue

* $_{p,i}= P_{nom,i} * f_{l,i} * _{i} $

```python
# Visualization of Probability Distributions for M/M/, Mn/M/, and Mx/M/ Models

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm

# Parameters
lambda_h = 15  # Average arrival rate (jobs/second)
mu_h = 20       # Service rate (jobs/second)
batch_mean = 8  # Mean batch size for Mx/M/ model
batch_std = 2   # Standard deviation of batch size

# Calculate the offered load ( = /)
offered_load = lambda_h / mu_h

# States (power levels in kW)
states = np.arange(0, 70, 1)

# 1. M/M/ Model: Poisson Distribution
p_mm_inf = poisson.pmf(states, offered_load)

# 2. Mn/M/ Model: Batch arrivals with constant size
batch_size_constant = 3
p_mn_inf = poisson.pmf(states / batch_size_constant, offered_load / batch_size_constant)

# 3. Mx/M/ Model: Batch arrivals with variable size (Normal Distribution)
batch_sizes = np.arange(1, 10)
batch_probs = norm.pdf(batch_sizes, loc=batch_mean, scale=batch_std)
batch_probs /= np.sum(batch_probs)  # Normalize to sum to 1

p_mx_inf = np.zeros_like(states, dtype=float)
for b_size, b_prob in zip(batch_sizes, batch_probs):
    p_mx_inf += b_prob * poisson.pmf(states / b_size, offered_load / b_size)

# Normalize probabilities for better visualization
p_mm_inf /= np.sum(p_mm_inf)
p_mn_inf /= np.sum(p_mn_inf)
p_mx_inf /= np.sum(p_mx_inf)


# Plot the distributions
plt.figure(figsize=(12, 8))
plt.plot(states, p_mm_inf, label="M/M/ Model", color="blue", linestyle=":")
plt.plot(states, p_mn_inf, label="Mn/M/ Model", color="green", linestyle="--")
plt.plot(states, p_mx_inf, label="Mx/M/ Model", color="red", linestyle=":")
plt.title("Probability Distributions for M/M/, Mn/M/, and Mx/M/ Models", fontsize=14)
plt.xlabel("States (kW)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.show()
```
